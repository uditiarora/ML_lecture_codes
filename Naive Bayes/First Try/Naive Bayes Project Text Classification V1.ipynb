```python
import operator as op
```

```python
#extracting words from the document
#first loop over each folder of the list of foldes of documents
#then iterate over each document in that folder
#then reading each token of the entered document then adding that token as a word into the dictionary of the words only if they satisfy 
#3 conditons 
#1-token comprose of all alphabets characters(got rid of the metadata this way)
#2-the length of the token is greater than 1(words with only 1 alphabet are omitted)
#3-the token is not a stop word
```

```python
dictionary={}

for folder in ls:
    files=os.listdir('./20_newsgroups/'+folder)  # accessing inner folders in all types of news
    for file in files:
        f=open('./20_newsgroups/'+folder+'/'+file,'r')   # accessing files in inner folders
        textfile=f.read()
        tokens=re.compile('\w+').findall(textfile)      # making object of re
        
        for token in tokens:
            if(token.isalpha() and len(token)>2 and token.lower() not in stop_words):
                if(token.lower() not in dictionary.keys()):
                    dictionary[token.lower()]=1
                else:
                    dictionary[token.lower()]+=1
            
listoftuples=sorted(dictionary.items(),key=op.itemgetter(1))
```

Loop through each folder of documents and then loop through each document in that folder and then reaching to each token as done in th process of making the dictionary of words then after finding the list of valid words from each document. for each feature word in the final_vocab dictionary find the count of that feature in the extracted word list from the document. Then appending that count to the row list and completing that row appending the type of document it belongs to and finally adding that row to that dataframe

```python
working_iterator=0
for folder in ls:
    files=os.listdir('./20_newsgroups/'+folder)
    #print(len(files))
    for file in files:
        words=[]
        count=0
        current_count=0
        f=open('./20_newsgroups/'+folder+'/'+file,'r')
        textfile=f.read()
        tokens=re.compile('\w+').findall(textfile)
        
        for token in tokens:
            if(token.isalpha() and len(token)>2 and token.lower() not in stop_words):
                words.append(token.lower())                     # words appended in words list as many times as they occur
                
        row=[]
        for feature in final_vocab.keys():
            for word in words:
                if(feature==word):
                    count+=1
                    
            row.append(count)  # appending count of each word in row
            count=0
            
        row.append(folder)  # adding type of news group at the end
        print(working_iterator)
        working_iterator+=1
        #print(i,":",len(row))
        df.loc[len(df)]=row
```
