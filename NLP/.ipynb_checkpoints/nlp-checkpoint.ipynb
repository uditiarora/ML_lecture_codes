```python
text_data = data["text"].copy()
text_data[0]
```

```python
print(text_data[0])
for i in range(len(text_data)):
    string = ""
    for words in text_data[i].split():
        if words[0]!='@':
            string += (words+" ")
    text_data[i] = string
text_data[0]
```

```python
punctuations = list(string.punctuation)
stop = stopwords.words('english')
stop += punctuations
```

```python
def get_simple_pos(tag):
    if(tag.startswith('J')):
        return wordnet.ADJ
    elif(tag.startswith('V')):
        return wordnet.VERB
    elif(tag.startswith('N')):
        return wordnet.NOUN
    elif(tag.startswith('R')):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
#filtering
lemmatizer = WordNetLemmatizer()
for i in range(len(text_data)):
    li = word_tokenize(text_data[i])
    li = [w for w in li if not w.lower() in stop]
    for j in range(len(li)):
        pos = pos_tag([li[j]])
        print(li[j],end=" ")
        li[j] = lemmatizer.lemmatize(li[j],pos = get_simple_pos(pos[0][1]))
        print(li[j])
    text_data[i] = " ".join(li)
text_data[0]
```
